## Part 2: Data Summarization

```python
# Total "Gross" of all top 1000 movies
total_gross = imdb['Gross'].sum()

total_gross
```

```python
# Average "No_of_Votes" of all movies
avg_votes = imdb['No_of_Votes'].mean()

avg_votes
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order
decade_counts = imdb['Decade'].value_counts()

decade_counts
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list
top_10_directors = imdb['Director'].value_counts().head(10)

top_10_directors
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross
director_stats = imdb.groupby('Director').agg(
    Number_of_Movies=('Series_Title', 'count'),
    Avg_IMDB_Rating=('IMDB_Rating', 'mean'),
    Total_Gross=('Gross', 'sum')
)

top_10_gross = director_stats.sort_values(by='Total_Gross', ascending=False).head(10)

top_10_gross
```

```python
# Group movies by decade and director
# In each group (i.e., for each decade and each director),
# calculate the number of movies and average IMDB rating.
# Sort in descending order of movie count
decade_dir = imdb.groupby(['Decade', 'Director']).agg(
    Number_of_Movies=('Series_Title', 'count'),
    Avg_IMDB_Rating=('IMDB_Rating', 'mean')
)

decade_dir = decade_dir.sort_values(by='Number_of_Movies', ascending=False)

decade_dir.head(10)
```

```python
# Find the top 3 highest rated movies of each year since 2010.
# Show their released year, ranking, title, and IMDB_Rating.
# Sort them in descending order of year and ascending order of ranking
recent = imdb[imdb['Released_Year'] >= 2010].copy()

recent = recent.sort_values(by=['Released_Year', 'IMDB_Rating'],
                                          ascending=[False, False])

recent['ranking'] = recent.groupby('Released_Year').cumcount() + 1

top_3 = recent[recent['ranking'] <= 3]
result = top_3[['Released_Year', 'ranking', 'Series_Title', 'IMDB_Rating']]

result = result.sort_values(by=['Released_Year', 'ranking'], ascending=[False, True])

result
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.
dir_decade_stats = imdb.groupby(['Director', 'Decade']).size().reset_index(name='Movie_Count')

decade_counts = dir_decade_stats.groupby('Director')['Decade'].nunique()
career_directors = decade_counts[decade_counts >= 5].index

big_directors = dir_decade_stats[dir_decade_stats['Director'].isin(career_directors)]

big_directors = big_directors.sort_values(by=['Director', 'Decade'])

big_directors
```
