## Question 1. Selecting rows

From the 'flights' dataframe, find all flights that satisfy the following certain conditions:

## Question 4. Adding new columns

Currently dep_time and sched_dep_time are convenient to look at, but hard to compute with because they’re not really continuous numbers.

For example, 759 means 7:59 and 801 means 8:01. Their difference is not 42 but 2 minutes.

```python
# Convert them to a more convenient representation of number of minutes since midnight (0).
flights['dep_time_mins'] = (flights['dep_time'] // 100 * 60 + flights['dep_time'] % 100) % 1440
flights['sched_dep_time_mins'] = (flights['sched_dep_time'] // 100 * 60 + flights['sched_dep_time'] % 100) % 1440
```

```python
# Create a new column of arr_time - dep_time.


# Compare this column with air_time.
flights['arr_time_mins'] = (flights['arr_time'] // 100 * 60 + flights['arr_time'] % 100) % 1440

flights['duration'] = (flights['arr_time_mins'] - flights['dep_time_mins']) % 1440

flights[['duration', 'air_time']].head()
```

```python
# Compare dep_time, sched_dep_time, and dep_delay. How would you expect those three numbers to be related?
# Try creating a column to calculate dep_delay from dep_time and sched_dep_time (and/or other columns if necessary).
dep_mins = (flights['dep_time'] // 100 * 60 + flights['dep_time'] % 100)
sched_mins = (flights['sched_dep_time'] // 100 * 60 + flights['sched_dep_time'] % 100)

flights['calc_delay'] = dep_mins - sched_mins

flights.loc[flights['calc_delay'] < -1200, 'calc_delay'] += 1440
flights.loc[flights['calc_delay'] > 1200, 'calc_delay'] -= 1440

# Test your results.
comparison = flights[['dep_time', 'sched_dep_time', 'dep_delay', 'calc_delay']]

mismatches = comparison[comparison['dep_delay'] != comparison['calc_delay']]

print(f"Number of mismatches: {len(mismatches)}")
mismatches.head()
```

```python
# Find the 20 most delayed flights.
# Display the following: year,month,day,carrier,flight,dep_delay,arr_delay,carrier
# How do you want to handle ties?
columns = ['year', 'month', 'day', 'carrier', 'flight', 'dep_delay', 'arr_delay']

top_20_nlargest = flights.nlargest(20, 'dep_delay', keep='first')

top_20_nlargest[columns]
```

```python
# Sort all AA flights to find the top 10 fastest (highest speed) flights.
# Display the following: year,month,day,carrier,flight,orig,dest,distance,air_time,speed (miles per hour)
AA_flights = flights[flights['carrier'] == 'AA'].copy()

AA_flights['speed'] = (AA_flights['distance'] / AA_flights['air_time']) * 60

fastest_aa = AA_flights.sort_values(by='speed', ascending=False).head(10)

columns = ['year', 'month', 'day', 'carrier', 'flight', 'origin', 'dest', 'distance', 'air_time', 'speed']
fastest_aa[columns]
```

```python
# Find all flights that satisfy the following:
# - From John F. Kennedy Airpot (JFK) or Newark Aiport (EWR) to Seattle-Tacoma Airport (SEA)
# - Carrier is UA, AA, or DL.
# - Dates from 4/1/2013 (inclusive) to 4/3/2013 (inclusive)
# - Scheduled arrival time is before noon.
# - Display the following: year,month,day,carrier,flight,origin,dest,sched_dep_time,sched_arr_time
# - Sort by year, month, day, sched_arr_time
filtered_flights = flights[
    (flights['origin'].isin(['JFK', 'EWR'])) &
    (flights['dest'] == 'SEA') &
    (flights['carrier'].isin(['UA', 'AA', 'DL'])) &
    (flights['year'] == 2013) &
    (flights['month'] == 4) &
    (flights['day'].between(1, 3)) &
    (flights['sched_arr_time'] < 1200)
].copy()

sorted_flights = filtered_flights.sort_values(
    by=['year', 'month', 'day', 'sched_arr_time']
)

cols = ['year', 'month', 'day', 'carrier', 'flight', 'origin', 'dest', 'sched_dep_time', 'sched_arr_time']
sorted_flights[cols]
```
